```python
# YOLO 설치 및 초기 설정 가이드

"""
1. 필수 라이브러리 설치
명령 프롬프트(CMD)나 터미널에서 다음 명령어들을 순서대로 실행하세요.
"""

# ================== 설치 명령어들 ==================
# 다음 명령어들을 CMD/터미널에서 실행하세요:

# 1. 기본 패키지 업그레이드
# pip install --upgrade pip

# 2. PyTorch 설치 (GPU 버전 - CUDA 사용 가능한 경우)
# pip install torch torchvision torchaudio --index-url https://download.pytorch.org/whl/cu118

# 3. PyTorch 설치 (CPU 버전만 필요한 경우)
# pip install torch torchvision torchaudio

# 4. Ultralytics YOLO 설치
# pip install ultralytics

# 5. 기타 필요한 라이브러리들
# pip install opencv-python matplotlib pillow numpy pandas

# ================== 설치 확인 코드 ==================

import sys
import subprocess
import importlib.util

def check_package_installed(package_name):
    """패키지 설치 여부 확인"""
    spec = importlib.util.find_spec(package_name)
    return spec is not None

def install_package(package_name):
    """패키지 설치"""
    try:
        subprocess.check_call([sys.executable, "-m", "pip", "install", package_name])
        print(f"✅ {package_name} 설치 완료")
        return True
    except subprocess.CalledProcessError:
        print(f"❌ {package_name} 설치 실패")
        return False

def setup_yolo_environment():
    """YOLO 환경 자동 설정"""
    print("🚀 YOLO 환경 설정을 시작합니다...\n")
    
    # 필수 패키지 목록
    required_packages = {
        'torch': 'torch',
        'torchvision': 'torchvision', 
        'ultralytics': 'ultralytics',
        'cv2': 'opencv-python',
        'matplotlib': 'matplotlib',
        'PIL': 'pillow',
        'numpy': 'numpy',
        'pandas': 'pandas'
    }
    
    # 패키지 설치 확인 및 설치
    for import_name, package_name in required_packages.items():
        if check_package_installed(import_name):
            print(f"✅ {package_name} 이미 설치됨")
        else:
            print(f"📦 {package_name} 설치 중...")
            install_package(package_name)
    
    print("\n🎉 모든 패키지 설치가 완료되었습니다!")

# ================== YOLO 모델 다운로드 및 테스트 ==================

def download_yolo_models():
    """YOLO 모델들을 미리 다운로드"""
    try:
        from ultralytics import YOLO
        
        print("📥 YOLO 모델을 다운로드합니다...")
        
        # 다양한 YOLO 모델 다운로드
        models = {
            'yolov8n.pt': 'YOLOv8 Nano (가장 빠름, 가벼움)',
            'yolov8s.pt': 'YOLOv8 Small (균형)',
            'yolov8m.pt': 'YOLOv8 Medium (정확도 향상)',
            'yolov8l.pt': 'YOLOv8 Large (높은 정확도)',
            'yolov8x.pt': 'YOLOv8 Extra Large (최고 정확도)'
        }
        
        for model_name, description in models.items():
            try:
                print(f"다운로드 중: {model_name} - {description}")
                model = YOLO(model_name)
                print(f"✅ {model_name} 다운로드 완료")
            except Exception as e:
                print(f"❌ {model_name} 다운로드 실패: {e}")
        
        print("\n🎉 모델 다운로드가 완료되었습니다!")
        return True
        
    except ImportError:
        print("❌ ultralytics 패키지가 설치되지 않았습니다.")
        return False

def test_yolo_installation():
    """YOLO 설치 테스트"""
    try:
        from ultralytics import YOLO
        import cv2
        import torch
        
        print("🧪 YOLO 설치 테스트를 시작합니다...\n")
        
        # PyTorch 버전 확인
        print(f"PyTorch 버전: {torch.__version__}")
        print(f"CUDA 사용 가능: {torch.cuda.is_available()}")
        if torch.cuda.is_available():
            print(f"CUDA 버전: {torch.version.cuda}")
            print(f"GPU 개수: {torch.cuda.device_count()}")
        
        # OpenCV 버전 확인
        print(f"OpenCV 버전: {cv2.__version__}")
        
        # YOLO 모델 로드 테스트
        print("\n📥 YOLO 모델 로드 테스트...")
        model = YOLO('yolov8n.pt')  # 가장 작은 모델로 테스트
        print("✅ YOLO 모델 로드 성공!")
        
        # 더미 이미지로 추론 테스트
        import numpy as np
        dummy_image = np.random.randint(0, 255, (640, 640, 3), dtype=np.uint8)
        results = model(dummy_image)
        print("✅ YOLO 추론 테스트 성공!")
        
        print("\n🎉 모든 테스트가 성공적으로 완료되었습니다!")
        print("YOLO를 사용할 준비가 되었습니다! 🚀")
        return True
        
    except Exception as e:
        print(f"❌ 테스트 실패: {e}")
        return False

# ================== 기본 YOLO 사용 예제 ==================

def basic_yolo_example():
    """기본 YOLO 사용 예제"""
    try:
        from ultralytics import YOLO
        import cv2
        import numpy as np
        
        print("📖 기본 YOLO 사용 예제")
        print("=" * 50)
        
        # 1. 모델 로드
        model = YOLO('yolov8n.pt')
        print("✅ 모델 로드 완료")
        
        # 2. 예시 이미지 생성 (실제로는 파일에서 로드)
        # img = cv2.imread('your_image.jpg')  # 실제 사용시
        img = np.random.randint(0, 255, (640, 640, 3), dtype=np.uint8)
        
        # 3. 객체 감지 실행
        results = model(img)
        
        # 4. 결과 출력
        for result in results:
            boxes = result.boxes
            if boxes is not None:
                print(f"감지된 객체 수: {len(boxes)}")
                for box in boxes:
                    # 클래스 ID와 신뢰도
                    class_id = int(box.cls[0])
                    confidence = float(box.conf[0])
                    # 바운딩 박스 좌표
                    x1, y1, x2, y2 = box.xyxy[0].tolist()
                    
                    print(f"클래스 ID: {class_id}, 신뢰도: {confidence:.2f}")
                    print(f"좌표: ({x1:.0f}, {y1:.0f}, {x2:.0f}, {y2:.0f})")
        
        print("✅ 기본 예제 실행 완료!")
        
    except Exception as e:
        print(f"❌ 예제 실행 실패: {e}")

# ================== 커스텀 모델 학습 예제 ==================

def create_training_example():
    """커스텀 모델 학습을 위한 예제 코드"""
    training_code = '''
# 커스텀 YOLO 모델 학습 예제

from ultralytics import YOLO

def train_custom_model():
    """커스텀 데이터셋으로 YOLO 모델 학습"""
    
    # 1. 기본 모델 로드
    model = YOLO('yolov8n.pt')  # 사전 훈련된 모델에서 시작
    
    # 2. 데이터셋 설정 (YAML 파일 필요)
    # dataset.yaml 파일 예시:
    # train: /path/to/train/images
    # val: /path/to/val/images
    # nc: 1  # 클래스 개수
    # names: ['license_plate']  # 클래스 이름
    
    # 3. 모델 학습
    results = model.train(
        data='dataset.yaml',  # 데이터셋 설정 파일
        epochs=100,           # 학습 에포크 수
        imgsz=640,           # 이미지 크기
        batch=16,            # 배치 크기
        name='license_plate_detection'  # 실험 이름
    )
    
    # 4. 모델 평가
    results = model.val()
    
    # 5. 모델 내보내기
    model.export(format='onnx')  # ONNX 형식으로 내보내기

# 데이터셋 폴더 구조:
# dataset/
# ├── train/
# │   ├── images/
# │   └── labels/
# ├── val/
# │   ├── images/
# │   └── labels/
# └── dataset.yaml
'''
    
    print("📚 커스텀 모델 학습 코드 예제:")
    print("=" * 50)
    print(training_code)

# ================== 메인 실행 부분 ==================

if __name__ == "__main__":
    print("🎯 YOLO 설치 및 설정 도구")
    print("=" * 50)
    
    while True:
        print("\n선택하세요:")
        print("1. 환경 자동 설정 (패키지 설치)")
        print("2. YOLO 모델 다운로드")
        print("3. 설치 테스트")
        print("4. 기본 사용 예제")
        print("5. 커스텀 학습 예제 보기")
        print("6. 종료")
        
        choice = input("\n번호를 입력하세요: ").strip()
        
        if choice == '1':
            setup_yolo_environment()
        elif choice == '2':
            download_yolo_models()
        elif choice == '3':
            test_yolo_installation()
        elif choice == '4':
            basic_yolo_example()
        elif choice == '5':
            create_training_example()
        elif choice == '6':
            print("👋 프로그램을 종료합니다.")
            break
        else:
            print("❌ 잘못된 선택입니다. 다시 선택해주세요.")
```
